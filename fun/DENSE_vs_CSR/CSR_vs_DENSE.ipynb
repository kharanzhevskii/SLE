```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
CSR_fat_O0 = np.loadtxt(r'\\path\CSR_fat_O0.txt')
CSR_fat_O1 = np.loadtxt(r'\\path\CSR_fat_O1.txt')
CSR_fat_O2 = np.loadtxt(r'\\path\CSR_fat_O2.txt')
CSR_thin_O0 = np.loadtxt(r'\\path\CSR_thin_O0.txt')
CSR_thin_O1 = np.loadtxt(r'\\path\CSR_thin_O1.txt')
CSR_thin_O2 = np.loadtxt(r'\\path\CSR_thin_O2.txt')

dense_fat_O0 = np.loadtxt(r'\\path\dense_fat_O0.txt')
dense_fat_O1 = np.loadtxt(r'\\path\dense_fat_O1.txt')
dense_fat_O2 = np.loadtxt(r'\\path\dense_fat_O2.txt')
dense_thin_O0 = np.loadtxt(r'\\path\dense_thin_O0.txt')
dense_thin_O1 = np.loadtxt(r'\\path\dense_thin_O1.txt')
dense_thin_O2 = np.loadtxt(r'\\path\dense_thin_O2.txt')
```

```python
n = np.array([5*x for x in range(2, 1001)])
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.grid(linestyle='--')
plt.title("CSR * dense vector")
plt.xlabel(r"$n$", fontsize=12)
plt.ylabel(r"$t, nsec$", fontsize=12)

plt.scatter(n, CSR_fat_O0, color='#00BFFF', marker='.', s=10, label='O0')
plt.scatter(n, CSR_fat_O1, color='#1E90FF', marker='.', s=10, label='O1')
plt.scatter(n, CSR_fat_O2, color='#4B0082', marker='.', s=10, label='O2')

plt.legend()
plt.savefig(r'\\path\CSR_dense_Ocomp.png')
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.grid(linestyle='--')
plt.title("CSR * sparse vector")
plt.xlabel(r"$n$", fontsize=12)
plt.ylabel(r"$t, nsec$", fontsize=12)
plt.ylim(0, 160000)
plt.scatter(n, CSR_thin_O0, color='#00BFFF', marker='.', s=10, label='O0')
plt.scatter(n, CSR_thin_O1, color='#1E90FF', marker='.', s=10, label='O1')
plt.scatter(n, CSR_thin_O2, color='#4B0082', marker='.', s=10, label='O2')

plt.legend()
plt.savefig(r'\\path\CSR_sparse_Ocomp.png')
```

```python
n = np.array([5*x for x in range(2, 501)])
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.grid(linestyle='--')
plt.title("DENSE * dense vector")
plt.xlabel(r"$n$", fontsize=12)
plt.ylabel(r"$t, nsec$", fontsize=12)

plt.scatter(n, dense_fat_O0, color='#FFD700', marker='.', s=10, label='O0')
plt.scatter(n, dense_fat_O1, color='#FF8C00', marker='.', s=10, label='O1')
plt.scatter(n, dense_fat_O2, color='#FF4500', marker='.', s=10, label='O2')

plt.legend()
plt.savefig(r'\\path\DENSE_dense_Ocomp.png')
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.grid(linestyle='--')
plt.title("DENSE * sparse vector")
plt.xlabel(r"$n$", fontsize=12)
plt.ylabel(r"$t, nsec$", fontsize=12)

plt.scatter(n, dense_thin_O0, color='#FFD700', marker='.', s=10, label='O0')
plt.scatter(n, dense_thin_O1, color='#FF8C00', marker='.', s=10, label='O1')
plt.scatter(n, dense_thin_O2, color='#FF4500', marker='.', s=10, label='O2')

plt.legend()
plt.savefig(r'\\path\DENSE_sparse_Ocomp.png')
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.grid(linestyle='--')
plt.title("DENSE vs CSR * dense vector")
plt.xlabel(r"$n$", fontsize=12)
plt.ylabel(r"$t, nsec$", fontsize=12)

plt.scatter(n, CSR_fat_O2[:499], color='#4B0082', marker='.', s=10, label='CSR O2')
plt.scatter(n, dense_fat_O2, color='#FFD700', marker='.', s=10, label='Dense O2')

b, c, d = np.polyfit(n[50:], dense_fat_O2[50:], deg=2)
x_mm = np.linspace(500, 2500, 100)
plt.plot(x_mm, b*x_mm**2 + c*x_mm + d, color='#FF4500', linestyle='-.', linewidth=1.5, label='$O(n^2)$')

plt.legend()
plt.savefig(r'\\path\DENSE_vs_CSR_dense.png')
```

```python
plt.figure(figsize=(8, 6), dpi=200)
plt.grid(linestyle='--')
plt.title("DENSE vs CSR * sparse vector")
plt.xlabel(r"$n$", fontsize=12)
plt.ylabel(r"$t, nsec$", fontsize=12)

plt.scatter(n, CSR_thin_O2[:499], color='#4B0082', marker='.', s=10, label='CSR O2')
plt.scatter(n, dense_thin_O2, color='#FFD700', marker='.', s=10, label='Dense O2')

b, c, d = np.polyfit(n[50:], dense_fat_O2[50:], deg=2)
x_mm = np.linspace(500, 2500, 100)
plt.plot(x_mm, b*x_mm**2 + c*x_mm + d, color='#FF4500', linestyle='-.', linewidth=1.5, label='$O(n^2)$')

plt.legend()
plt.savefig(r'\\path\DENSE_vs_CSR_sparse.png')
```
